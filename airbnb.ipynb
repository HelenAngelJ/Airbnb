```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Reading the json file and setting the display option to view the entire column name


df = pd.read_json('sample_airbnb.json')
pd.set_option('display.max_columns',None)
df.head(3)
```

```python
address_df = pd.json_normalize(df['address'])
address_df.head()
```

```python
# getting the host detials
host_df = pd.json_normalize(df['host'])
host_df.head()
```

```python
review_score = pd.json_normalize(df['review_scores'])
review_score.head()
```

```python
available = pd.json_normalize(df['availability'])
available.head()
```

```python
host = pd.concat([host_df,available],axis=1)
host.head()
```

```python
df_new = pd.concat([df,address_df,host_df['host_name'],review_score['review_scores_rating']],axis=1)
df_new.drop(['address','host','review_scores','reviews','images','availability','reviews_per_month'],axis=1,inplace=True)
df_new.head()
```

```python
df_new.drop(['last_scraped','calendar_last_scraped','first_review','last_review'],axis=1,inplace=True)
```
