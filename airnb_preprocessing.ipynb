```python
data= pd.concat([df[['_id', 'name', 'summary', 'space', 'description',
       'neighborhood_overview', 'notes','access',
       'house_rules', 'property_type', 'room_type', 'bed_type',
       'minimum_nights', 'maximum_nights', 'cancellation_policy',
       'accommodates', 'bedrooms', 'beds', 'number_of_reviews', 'bathrooms',
       'amenities', 'price', 'security_deposit', 'cleaning_fee',
       'extra_people', 'guests_included']],
          host_df[['host_id','host_name','host_response_rate','host_is_superhost','host_identity_verified','host_total_listings_count']],
          review_scores_df[['review_scores_cleanliness',
       'review_scores_checkin', 'review_scores_communication',
       'review_scores_location', 'review_scores_value',
       'review_scores_rating']],
       address_df[['street', 'suburb', 'government_area', 'market', 'country',
       'country_code', 'location.type', 'location.coordinates']],
       avail_df],axis=1
       )
```

```python
data['bedrooms'].fillna(0, inplace=True)
data['beds'].fillna(0, inplace=True)
data['bathrooms'].fillna(0, inplace=True)
data['security_deposit'].fillna(0,inplace=True)
data['cleaning_fee'].fillna(0, inplace=True)
data['review_scores_checkin'].fillna(data['review_scores_checkin'].mode().values[0],inplace=True)
data['review_scores_cleanliness'].fillna(data['review_scores_cleanliness'].mode().values[0],inplace=True)
data['review_scores_communication'].fillna(data['review_scores_communication'].mode().values[0],inplace=True)
data['review_scores_location'].fillna(data['review_scores_location'].mode().values[0],inplace=True)
data['review_scores_rating'].fillna(data['review_scores_rating'].mode().values[0],inplace=True)
data['review_scores_value'].fillna(data['review_scores_value'].mode().values[0],inplace=True)
data['host_response_rate'].fillna(data['host_response_rate'].mean(),inplace=True)
```

```python
data.isnull().sum()
```

```python
data['minimum_nights'] = data['minimum_nights'].astype('int64')
data['maximum_nights'] = data['maximum_nights'].astype('int64')
data['bedrooms'] = data['bedrooms'].astype('int64')
data['beds'] = data['beds'].astype('int64')
data['bathrooms'] = data['bathrooms'].astype(str).astype(float).astype('int64')
data['price'] = data['price'].astype(str).astype(float)
data['cleaning_fee'] = data['cleaning_fee'].astype(str).astype(float)
data['security_deposit'] = data['security_deposit'].astype(str).astype(float)
data['extra_people'] = data['extra_people'].astype(str).astype(float).astype('int64')
data['guests_included'] = data['guests_included'].astype(str).astype('int64')
data['review_scores_checkin'] = data['review_scores_checkin'].astype('int64')
data['review_scores_cleanliness'] = data['review_scores_cleanliness'].astype('int64')
data['review_scores_communication'] = data['review_scores_communication'].astype('int64')
data['review_scores_location'] = data['review_scores_location'].astype('int64')
data['review_scores_rating'] = data['review_scores_rating'].astype('int64')
data['review_scores_value'] = data['review_scores_value'].astype('int64')
data['host_response_rate'] = data['host_response_rate'].astype('int64')
```
